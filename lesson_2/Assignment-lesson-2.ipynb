```python
import os
import re
from hanziconv import HanziConv
from collections import Counter
```

```python
title_pattern = re.compile('title="[\w]+"')
def getTitle(strLine:str):
    titles = []
    for subwords in strLine.split():
        match =  re.findall(title_pattern,subwords)
        if match:
            match = match[0].split('=')
            strTitle = match[1].strip('""')
            titles.append(HanziConv.toSimplified(strTitle))# + match
    return titles
```

```python

#  返回一个文件内的关键字，  title 需 转为简体中文
def getArticlesInFile(filename:str)->list:
    lstArticles = []
    try:
        file_obj = open(filename,"r")
        for line in file_obj:
            if line.startswith("<doc"):
                #print(line)
                lstArticles = lstArticles + getTitle(line) 
    except UnicodeDecodeError:
        return lstArticles
    return lstArticles
    
```

```python
def getFilesInPath(strPath:str)->list:
    ret_files = []
    lstFiles = os.listdir(strPath)
    for i in range(len(lstFiles)):
        filepath = os.path.join(strPath,lstFiles[i])
        if os.path.isdir(filepath):
            ret_files.extend(getFilesInPath(filepath))
        else:
            ret_files.append(filepath)
    return ret_files
```

```python

def getArticlesInDir(strPath:str)->list:
    articles = []
    files = getFilesInPath(strPath)
    for file in files:
        articles.extend(getArticlesInFile(file))
    return articles
```

```python
words_count = Counter(allArticles)
```

```python
import jieba
def cut(string):return list(jieba.cut(string))
```

```python
# about wordCounts
def get_prob(word):
    esp = 1/articleCounts
    if word in words_count:
        return words_count[word] /articleCounts
    else:
        return esp
```

```python
valida_tokens = [str(t) for t in allArticles]
```

```python
all_2_grams_words = [''.join(valida_tokens[i:i+2]) for i in range(len(valida_tokens[:-2]))]
```

```python
from collections import Counter
_2_gram_sum = len(all_2_grams_words)
_2_gram_counter = Counter(all_2_grams_words)
```

```python
def get_combination_prob(w1,w2):
    if w1 + w2 in _2_gram_counter:
        return _2_gram_counter[w1+w2]/_2_gram_sum
    else:
        return 1/_2_gram_sum
```

```python
def get_prob_2_gram(w1, w2):
    return get_combination_prob(w1, w2) / get_prob(w1)
```

```python
def langauge_model_of_2_gram(sentence):
    sentence_probability = 1
    
    words = cut(sentence)
    
    for i, word in enumerate(words):
        if i == 0: 
            prob = get_prob(word)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous, word)
        sentence_probability *= prob
    
    return sentence_probability
```
